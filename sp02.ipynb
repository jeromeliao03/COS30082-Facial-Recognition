```python
import os
import random
import shutil
from pathlib import Path

# Original dataset paths
train_original = Path("/Users/lasithcharuka/software/ml2/movlas/movlas2/fit/LCC_FASD 2/LCC_FASD_training")
val_original = Path("/Users/lasithcharuka/software/ml2/movlas/movlas2/fit/LCC_FASD 2/LCC_FASD_evaluation")

# New prepared dataset path
output_dir = Path("anti_spoof_prepared")

# New folder structure
folders = [
    output_dir / "train" / "real",
    output_dir / "train" / "spoof",
    output_dir / "val" / "real",
    output_dir / "val" / "spoof"
]

for folder in folders:
    folder.mkdir(parents=True, exist_ok=True)

# Image extensions
image_exts = [".jpg", ".jpeg", ".png"]

def get_images(folder):
    return [
        file for file in folder.iterdir()
        if file.suffix.lower() in image_exts
    ]

# Load original images
train_real = get_images(train_original / "real")
train_spoof = get_images(train_original / "spoof")

val_real = get_images(val_original / "real")
val_spoof = get_images(val_original / "spoof")

print("Original counts:")
print("Train real:", len(train_real))
print("Train spoof:", len(train_spoof))
print("Val real:", len(val_real))
print("Val spoof:", len(val_spoof))

# Balance training set
train_size = min(len(train_real), len(train_spoof))
val_size = min(len(val_real), len(val_spoof))

random.seed(42)

train_real_sample = random.sample(train_real, train_size)
train_spoof_sample = random.sample(train_spoof, train_size)

val_real_sample = random.sample(val_real, val_size)
val_spoof_sample = random.sample(val_spoof, val_size)

def copy_images(files, destination):
    for file in files:
        shutil.copy(file, destination / file.name)

# Copy balanced images
copy_images(train_real_sample, output_dir / "train" / "real")
copy_images(train_spoof_sample, output_dir / "train" / "spoof")

copy_images(val_real_sample, output_dir / "val" / "real")
copy_images(val_spoof_sample, output_dir / "val" / "spoof")

print("\nPrepared dataset created successfully.")
print("Train real:", len(os.listdir(output_dir / "train" / "real")))
print("Train spoof:", len(os.listdir(output_dir / "train" / "spoof")))
print("Val real:", len(os.listdir(output_dir / "val" / "real")))
print("Val spoof:", len(os.listdir(output_dir / "val" / "spoof")))
```
