## 2 - Data Setup

Data is excluded since it is very large. These steps automate, and standardise the setup.

Easiest way is honestly just downloading the zip from [here](https://www.kaggle.com/competitions/11-785-fall-20-homework-2-part-2/data) and placing it in the project root, and then issuing the following commands.

**Extract**

````python
import subprocess, os

archive = '../11-785-fall-20-homework-2-part-2.zip'
data_dir = '../data'

if not os.path.exists(archive):
    print(f"ERROR: archive not found at {archive}")
else:
    os.makedirs(data_dir, exist_ok=True)
    subprocess.run(['unzip', '-q', archive, '-d', data_dir], check=True)
    print(f"Extracted to {data_dir}")
````

**Delete zip**

````python
if os.path.exists(archive):
    os.remove(archive)
    print(f"Deleted {archive}")
else:
    print("Archive not found, skipping")
````
